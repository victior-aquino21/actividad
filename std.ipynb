```python
import pandas as pd
import random

# =========================
# 1. DATASET
# =========================

nombres = ["Juan", "Maria", "Carlos", "Ana", "Luis", "Pedro", "Sofia", "Elena"]
apellidos = ["Perez", "Gomez", "Lopez", "Ramirez", "Torres", "Diaz"]
roles = ["Profesor", "Trabajador"]
departamentos = ["Matematicas", "Ciencias", "Administracion", "Recursos Humanos", "Lenguaje"]

data = []

for i in range(200):
    nombre = random.choice(nombres) + " " + random.choice(apellidos)
    edad = random.randint(22, 65)
    rol = random.choice(roles)
    salario = random.randint(2500, 12000)
    departamento = random.choice(departamentos)
    años_exp = random.randint(1, 40)

    data.append([nombre, edad, rol, salario, departamento, años_exp])

df = pd.DataFrame(data, columns=[
    "Nombre", "Edad", "Rol", "Salario", "Departamento", "Años_Experiencia"
])

print("=== DATASET COMPLETO (primeras 10 filas) ===")
display(df.head(10))


# =========================
# 2. USO DE .loc
# =========================

print("\n=== SOLO PROFESORES ===")
display(df.loc[df["Rol"] == "Profesor"].head())

print("\n=== PERSONAS MAYORES DE 40 AÑOS ===")
display(df.loc[df["Edad"] > 40].head())

print("\n=== SOLO COLUMNAS NOMBRE Y SALARIO ===")
display(df.loc[:, ["Nombre", "Salario"]].head())

# Modificar un valor
df.loc[0, "Salario"] = 15000
print("\n=== FILA 0 MODIFICADA (.loc) ===")
display(df.loc[[0]])


# =========================
# 3. USO DE .iloc
# =========================

print("\n=== PRIMERAS 5 FILAS ===")
display(df.iloc[0:5])

print("\n=== FILAS 0-9 Y COLUMNAS 0-2 ===")
display(df.iloc[0:10, 0:3])

print("\n=== VALOR ESPECÍFICO (fila 0, columna 2) ===")
print(df.iloc[0, 2])

print("\n=== ÚLTIMAS 5 FILAS ===")
display(df.iloc[-5:])


# =========================
# 4. GUARDAR ARCHIVO
# =========================

df.to_csv("dataset_profesores_trabajadores.csv", index=False)

print("\nDataset guardado como 'dataset_profesores_trabajadores.csv'")
```
