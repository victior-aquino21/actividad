```python
import pandas as pd
import random
import matplotlib.pyplot as plt
from io import StringIO

# =========================
# 1. DATASET 
# =========================

platos = ["Pizza", "Hamburguesa", "Pasta", "Ensalada", "Tacos", "Sushi"]
categorias = ["Comida", "Bebida", "Postre"]

csv_data = "Orden_ID,Plato,Categoria,Precio,Cantidad,Total\n"

for i in range(1000):
    orden_id = i + 1
    plato = random.choice(platos)
    categoria = random.choice(categorias)
    precio = round(random.uniform(5, 25), 2)
    cantidad = random.randint(1, 5)
    total = round(precio * cantidad, 2)

    csv_data += f"{orden_id},{plato},{categoria},{precio},{cantidad},{total}\n"

# Convertir texto CSV a DataFrame
df = pd.read_csv(StringIO(csv_data))

print("=== DATASET (primeras 10 filas) ===")
display(df.head(10))


# =========================
# 2. ESTADÍSTICAS
# =========================

print("\n=== MEDIA ===")
print(df[["Precio", "Cantidad", "Total"]].mean())

print("\n=== MÍNIMO ===")
print(df[["Precio", "Cantidad", "Total"]].min())

print("\n=== MÁXIMO ===")
print(df[["Precio", "Cantidad", "Total"]].max())

print("\n=== DESVIACIÓN ESTÁNDAR ===")
print(df[["Precio", "Cantidad", "Total"]].std())


# =========================
# 3. GRÁFICA DE BARRAS
# =========================

ventas_por_plato = df.groupby("Plato")["Total"].sum()

plt.figure()
ventas_por_plato.plot(kind="bar")

plt.title("Ventas Totales por Plato")
plt.xlabel("Plato")
plt.ylabel("Total de Ventas")

plt.show()

df.to_csv("dataset_restaurante.csv", index=False)

print("\nDataset guardado como 'dataset_restaurante.csv'")
```
